```python
import opendatasets as od
import pandas as pd
```

```python
# Read in the sp500sub.csv that was generated earlier
df = pd.read_csv('./sp500sub.csv', index_col='Date', parse_dates=True)
```

```python
# Show intial few rows
df.head()
```

```python
# Show close- and adjusted close prices for Google.
# We can see that the 'close' prices have been adjusted to stock splits, as the Google stock was split in 2014.
# Another thing we can observe is that Google does not pay dividends. Otherwise the 'close' and 'adjusted close'
# prices would be different
google = df[df['Name'] == 'GOOG']
google[['Close', 'Adj Close']].plot(figsize=(10, 10))
```

```python
# Download the S&P500 data from Kaggle
od.download("https://www.kaggle.com/datasets/camnugent/sandp500/download?datasetVersionNumber=4")
df2 = pd.read_csv('./sandp500/all_stocks_5yr.csv', index_col='date', parse_dates=True)
df2.head()
```

```python
# Plot close prices for Google. We can see that the close prices have been adjusted for stock splits.
# Google stock was split in 2014.
google2 = df2[df2['Name'] == 'GOOGL']
google2[['close']].plot(figsize=(10,10))
```

```python
# Plot 'close' and 'adjusted close' prices for Apple. This time we see that these are different.
# The close price has been adjusted for stock split that happened in 2014. However, since Apple
# pays dividends, the 'close' and 'adjusted close' prices are different.
apple = df[df['Name'] == 'AAPL']
apple[['Close', 'Adj Close']].plot(figsize=(10, 10))
```
